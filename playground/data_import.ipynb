```python
def load_course_statistics_type(path):
    statistics_type = spark.read.load(path,
                         format="csv", sep=",", inferSchema="true", 
                         header="true")\
        .withColumn("updated_at", col("updated_at").cast(TimestampType()))
    statistics_type.createOrReplaceTempView("statistics_type")
    return statistics_type

def load_course_statistics(path):
    course_statistics = spark.read.load(path,
                         format="csv", sep=",", inferSchema="true", 
                         header="true")\
        .withColumn("updated_at", col("updated_at").cast(TimestampType()))
    course_statistics.createOrReplaceTempView("course_statistics")
    return course_statistics

statistics = load_course_statistics("/Users/LTV/Documents/school/code/data/db_data/educational_course_statistic_16.11.csv")
statistics_type = load_course_statistics_type("/Users/LTV/Documents/school/code/data/db_data/statistic_type_16.11.csv")
```

```python
spark.sql(
    """
    with statistics as (
        SELECT 
        profile_id, statistics_type.entity_id as stat_type, 
        course_statistics.created_at as created_at, educational_course_id
        FROM course_statistics
        JOIN statistics_type on course_statistics.statistic_type_id = statistics_type.id
        WHERE profile_id IS NOT NULL and educational_course_id IS NOT NULL and (entity_id = 2 or entity_id = 3)
        LIMIT 1000
    ),
    statistics_ends_labeled as (
        SELECT
        profile_id, educational_course_id, stat_type,
        IF(
            LAG(educational_course_id, 1, NULL) OVER (
                ORDER BY profile_id, educational_course_id, created_at
            ) != educational_course_id, created_at, NULL
        ) as start,
        IF(
            LEAD(educational_course_id, 1, NULL) OVER (
                ORDER BY profile_id, educational_course_id, created_at
            ) != educational_course_id, created_at, NULL
        ) as end,
        created_at
        FROM statistics
    ),
    session_duration as (
        SELECT 
        profile_id, educational_course_id, MIN(start) as started_at, MAX(END) as ended_at
        FROM
        statistics_ends_labeled
        GROUP BY profile_id, educational_course_id
    ) 
    SELECT * FROM session_duration WHERE started_at IS NOT NULL and ended_at IS NOT NULL
    """
).limit(30).toPandas()
```
